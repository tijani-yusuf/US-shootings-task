```python
#I want to import required packages to get my bar chart
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
plt.figure(figsize=(9,5))
plt.figure.facecolor = '#00000000'
sns.barplot(year_most_killed.index, year_most_killed,)
plt.title("The number killed by year")
plt.ylabel('number_of deaths')
plt.xlabel('year_most_killed')
```

```python
#the range of ages that were shot the most
#plotting a histogram to know the range of ages that were killed mostly
#the bin for the histogram np.arange(5,100,5)
#that shows the starting point, the interval and stopt point
import numpy as np
plt.figure(figsize=(20, 10))
plt.title("Age shot most")
plt.xlabel('Age')
plt.ylabel('Number of deaths')
plt.hist(shootings.age, bins=np.arange(5,100,5), color='green');
```
